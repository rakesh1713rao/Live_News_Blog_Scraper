```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://timesofindia.indiatimes.com/world/middle-east/israel-iran-war-news-live-udates-conflict-attack-middle-east-state-of-emergency-explosions-tehran-gulf-countries-trump-us-ali-khamenei/liveblog/128933029.cms'
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com/"
}
```

```python
webpage = requests.get(URL,headers=HEADERS)
```

```python
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
def get_time_updates(soup):
    try:
        time_updates = soup.find("div", attrs={"class":"fshNz"}).text
        time_updates = time_updates.strip()
    except AttributeError:
        time_updates = ""
    return time_updates
```

```python
def get_title_updates(soup):
    try:
        title_updates = soup.find("div", attrs={"class":"AimyF"}).text
        title_updates = title_updates.strip()
    except AttributeError:
        title_updates = ""
    return title_updates
```

```python
def get_description_updates(soup):
    try:
        description_updates = soup.find("div", attrs={"class":"sL5kN"}).text
        description_updates = description_updates.strip()
    except AttributeError:
        description_updates = ""
    return description_updates
```

```python
links = soup.find_all("div", attrs={"class":"IDar7 p8L0K"})
```

```python
links
```

```python
new_dict = {"news_time":[], "news_update":[], "news_desc":[]}
```

```python
for i in links:

    new_dict["news_time"].append(get_time_updates(i))
    new_dict["news_update"].append(get_title_updates(i))
    new_dict["news_desc"].append(get_description_updates(i))

new_details_df = pd.DataFrame.from_dict(new_dict)
```
